```python
import importer
import requests
from langchain.chains.mapreduce import MapReduceChain
from langchain.text_splitter import CharacterTextSplitter
from langchain_extras.llms.exllama import Exllama, BasicStreamingHandler
from langchain.prompts import PromptTemplate
from exllama.model import ExLlamaConfig


model_directory = "../models/TheBloke_GPT4All-13B-Snoozy-SuperHOT-8K-GPTQ"

handler = BasicStreamingHandler()

# Import the necessary modules and define the required variables
llm = Exllama(
    model_path=model_directory,
    lora_path=None, 
    temperature=0.7,
    beams=1,
    beam_length=40,
    stop_sequences=["Human:", "User:", "AI:"],
    callbacks=[handler],
    verbose=False,
    max_seq_len = 4096,
    alpha_value=4.0,  # For use with any models
    compress_pos_emb=4.0,  # For use with superhot
    # set_auto_map = "3, 2" #Gpu split, this will split 3gigs/2gigs
)
text_splitter = CharacterTextSplitter()
```

```python
from bs4 import BeautifulSoup
from langchain.docstore.document import Document


page = requests.get("https://jalammar.github.io/illustrated-transformer/")
soup = BeautifulSoup(page.text, 'html.parser')

texts = text_splitter.split_text(soup.get_text())  # List of documents to be summarized

docs = [Document(page_content=t) for t in texts]

```

```python
from langchain.chains.summarize import load_summarize_chain

chain = load_summarize_chain(llm, chain_type="map_reduce")
handler.set_chain(chain.llm_chain)
chain.run(docs)
```
